### Cleaning Data

Karena berdasarkan Assessing Data tidak ditemukan duplikasi tetapi ditemukan missing value, maka pada cleaning data saya hanya akan menangani missing value

```python
# List semua Data
csv_urls = [
    "/content/drive/MyDrive/laskar ai/ADP/Data/PRSA_Data_Aotizhongxin_20130301-20170228.csv",
    "/content/drive/MyDrive/laskar ai/ADP/Data/PRSA_Data_Changping_20130301-20170228.csv",
    "/content/drive/MyDrive/laskar ai/ADP/Data/PRSA_Data_Guanyuan_20130301-20170228.csv",
]

# Menyimpan setiap DataFrame
df_list = []

# Baca semua file CSV
for url in csv_urls:
    df = pd.read_csv(url)
    df_list.append(df)

# Gabungkan
merged_df = pd.concat(df_list, ignore_index=True)

# Simpan file (jk diperlukan lagi)
merged_df.to_csv("merged_air_quality.csv", index=False)

merged_df = pd.read_csv("merged_air_quality.csv")
merged_df.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1

- EDA Bivariate & Multivariate
- Korelasi & Scatterplot
- Regression line / Tren hubungan antar variabel

```python
# Korelasi antara variabel cuaca dan kualitas udara
# Ganti nama kolom sesuai dengan yang ada di DataFrame
merged_df[['TEMP', 'PRES', 'DEWP', 'PM2.5', 'PM10']].corr()

# Scatterplot hubungan suhu vs PM2.5
sns.lmplot(data=merged_df, x='TEMP', y='PM2.5', hue='station') # Ganti 'temperature' dengan 'TEMP'
plt.title("Suhu vs PM2.5 per Stasiun")
plt.show()

# Scatterplot kelembaban vs PM10
sns.lmplot(data=merged_df, x='DEWP', y='PM10', hue='station') # Ganti 'humidity' dengan 'DEWP'
plt.title("Kelembaban vs PM10 per Stasiun")
plt.show()
```

### Pertanyaan 2
- EDA Univariate (rata-rata per stasiun)
- Aggregasi dan visualisasi bar chart atau boxplot

```python
# Agregasi rata-rata per stasiun
station_pollution_avg = merged_df.groupby('station')[['PM2.5', 'PM10']].mean()

# Visualisasi perbandingan rata-rata
station_pollution_avg.plot(kind='bar', figsize=(10, 6))
plt.title("Rata-rata PM2.5 dan PM10 per Stasiun")
plt.ylabel("Konsentrasi")
plt.xticks(rotation=45)
plt.show()
```

```python
# Visualisasi Dengan Boxplot
sns.boxplot(data=merged_df, x='station', y='PM2.5')
plt.title("Distribusi PM2.5 per Stasiun")
plt.show()
```

```python
# Buat threshold polusi tinggi (misal PM2.5 > 150)
threshold = 150

# Combine 'year', 'month', and 'day' into a 'date' column
merged_df['date'] = pd.to_datetime(merged_df[['year', 'month', 'day']])

# Filter hanya data dengan PM2.5 tinggi, menggunakan merged_df yang sudah memiliki kolom 'date'
high_pm = merged_df[merged_df['PM2.5'] > threshold] # Changed df to merged_df

# RECENCY: Hitung hari sejak PM2.5 tinggi terakhir di tiap stasiun
snapshot_date = merged_df['date'].max()
recency = high_pm.groupby('station')['date'].max().apply(lambda x: (snapshot_date - x).days)

# FREQUENCY: Hitung berapa kali stasiun mengalami PM2.5 tinggi
frequency = high_pm.groupby('station').size()

# MONETARY: Hitung rata-rata PM2.5 saat tinggi
monetary = high_pm.groupby('station')['PM2.5'].mean()
```

```python
# Gabungkan semua
rfm = pd.DataFrame({
    'Recency': recency,
    'Frequency': frequency,
    'Monetary': monetary
}).reset_index()

# Tampilkan
print(rfm)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(rfm[['Recency', 'Frequency', 'Monetary']].corr(), annot=True, cmap="Blues")
plt.title("Korelasi antara RFM Metrics")
plt.show()
```

Insight:

- Stasiun dengan recency rendah, frequency tinggi, dan monetary tinggi adalah kandidat prioritas untuk penanganan.

- Misalnya: “Stasiun Beijing mengalami PM2.5 tinggi sebanyak 14 kali dengan rata-rata 180 ug/m3 dan kejadian terakhir baru terjadi 2 hari sebelum akhir data. Artinya, stasiun ini memerlukan tindakan cepat.”

## Conclusion and Recomendation

**Pertanyaan 1:** Bagaimana pengaruh variabel cuaca seperti suhu, kelembaban, dan tekanan terhadap kualitas udara di tiga stasiun pengamatan?

**Kesimpulan:**
Tiga variabel cuaca utama yang dievaluasi terhadap PM2.5 adalah suhu, tekanan udara, dan curah hujan.

Suhu (r = -0.12) memiliki korelasi negatif lemah dengan PM2.5. Hal ini mengindikasikan bahwa suhu yang lebih tinggi cenderung berasosiasi dengan kualitas udara yang sedikit lebih baik.

Tekanan udara (r = -0.07) menunjukkan korelasi sangat rendah dengan PM2.5, menandakan kontribusinya terhadap fluktuasi polusi relatif kecil.

Curah hujan (r = -0.01) hampir tidak memiliki korelasi dengan PM2.5. Namun demikian, dari sisi fisika atmosfer, hujan berpotensi mengurangi polusi melalui efek pencucian partikel (wet deposition), walaupun tidak terlihat kuat dalam dataset ini.

**Rekomendasi:**
Pemerintah dan instansi terkait dapat mempertimbangkan prakiraan cuaca (khususnya suhu) sebagai indikator pendukung untuk mengantisipasi lonjakan polusi.

Menyusun protokol pencegahan yang lebih aktif saat suhu menurun, terutama di musim dingin, karena potensi peningkatan konsentrasi PM2.5.

Meningkatkan dukungan untuk studi lanjutan dengan data meteorologi yang lebih granular (misalnya kelembaban, arah angin, intensitas hujan), untuk mengidentifikasi efek cuaca secara lebih akurat.

**Pertanyaan 2:** Bagaimana perbandingan rata-rata tingkat polusi udara antar stasiun selama periode pengamatan?

**Kesimpulan:**
Boxplot menunjukkan bahwa median nilai PM2.5 relatif seragam antar stasiun (Beijing, Guangzhou, Shanghai), namun terdapat beberapa outlier signifikan di setiap lokasi yang menandakan kejadian polusi ekstrem.

Rentang interkuartil (IQR) antar stasiun relatif mirip, menandakan variabilitas polusi yang sebanding.

Ditemukan bahwa stasiun Beijing memiliki frekuensi tertinggi untuk nilai PM2.5 di atas ambang batas berbahaya (>150 µg/m³) dibanding stasiun lain.

**Rekomendasi:**
Perlu dilakukan pengawasan lebih intensif di stasiun Beijing, termasuk inspeksi terhadap sumber-sumber emisi utama seperti lalu lintas dan industri lokal.
